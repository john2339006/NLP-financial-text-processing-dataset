```python
# we looped through the file to split them by @ character

list_of_lists = []

with open(path) as f:
    for line in f:
        inner_list = [line.strip() for line in line.split('@')]
        list_of_lists.append(inner_list)
```

```python
# dataframe created on the stacked data and assigned it a variable df

df = pd.DataFrame(list_of_lists)
```

```python
# now define a function to convert the review column into lower case and remove punctiations as well

import re
def preprocessor(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)',text)
    text = (re.sub('[\W]+', ' ', text.lower()) + ' '.join(emoticons).replace('-', ''))
    return text
```

```python
# now apply it on the review column

df['review'] = df['review'].apply(preprocessor)
```

```python
# a sample view to make sure that changes took place

df.iloc[0, 0]
```

```python
# a function to tokenise the review column

from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
```

```python
# apply the function on the review column

df['review'] = df['review'].apply(lambda x: tokenizer.tokenize(x.lower()))
```

```python
# import the needed library which contains the stopwords
from nltk.corpus import stopwords
```

```python
# assign the stopwords to a variable stop
stop = set(stopwords.words('english'))
```

```python
# a function to remove stop words from the data

def remove_stop(text):
    word = [w for w in text if w not in stop]
    return word
```

```python
# apply the function to the column review

df['review'] = df['review'].apply(lambda x: remove_stop(x))
```

```python
# import the needed library to use wordnet

import nltk
nltk.download('wordnet')
```

```python
# import the steming libraries to be used

from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
```

```python
# initialise the lemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize(text):
    word = [lemmatizer.lemmatize(w) for w in text]
    return word
```

```python
# apply the lemmatizer function on the review column

df['review'] = df['review'].apply(lambda x: lemmatize(x))
```

```python
# initialise the porter and assign it to a function

porter = PorterStemmer()
def port(text):
    word_port = [porter.stem(i) for i in text]
    return word_port
```

```python
# assign the function to the review column

df['review'] = df['review'].apply(lambda x: port(x))
```

```python
# we now convert the sentiments into numeric for the machine to understand

from sklearn.preprocessing import LabelEncoder
```

```python
# now call the class encoder and assign it to a variable, then finally transform

le = LabelEncoder()
df['sentiments'] = le.fit_transform(df['sentiments'])
```

```python
# we convert the review into numeric for machine purpose

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# now call the class encoder and assign it to a variable, then finally transform

cv = CountVectorizer(lowercase = False,stop_words = 'english', min_df=2)
```

```python
df = pd.DataFrame(df)
```

```python
df['review'] = df['review'].apply(str)
```

```python
x = df['review'].values
y = df['sentiments'].values
```

```python
x = cv.fit_transform(x)
```

```python
# now scale the review encoding with Tfid

from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfid = TfidfTransformer()
```

```python
x_vector = tfid.fit_transform(x)
```

```python
# now import the class for splitting the data

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_vector, y, test_size = 0.2, random_state = 5, stratify = y) 
```

```python
#######################  first algorithm (naive_bayes) ############################

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
gnb = MultinomialNB()
```

```python
model = gnb.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
score = accuracy_score(y_test, pred)
```

```python
print('the accuracy of the algorithm is {}'.format(score))
```

```python
class_report = classification_report(y_test, pred)
print(class_report)
```

```python
##########################  second algorithm (SVM) ############################
```

```python
from sklearn.svm import SVC

svc = SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
```

```python
model2 = svc.fit(x_train, y_train)
```

```python
# predicting the final output

pred2 = model2.predict(x_test)
pred2 = pd.DataFrame(pred2)
pred2.head()
```

```python
pred2 = le.inverse_transform(pred2)
```

```python
pred2 = pd.DataFrame(pred2)
pred2.columns = ['target']
pred2.head()
```
